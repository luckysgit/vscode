```python
import ctypes

class meralist:
    def __init__(self, size=1, n=0):
        self.size = size
        self.n = n
        self.A = self.create_array(self.size)

    # create array: with c type 
    def create_array(self, csize):
        return (csize * ctypes.py_object)()
    
    # length of array
    def __len__(self):
        return self.n
    
    def append(self, item):
        if self.n == self.size:
            # resize if full
            self.resize(self.size * 2)
            
        # append
        self.A[self.n] = item
        self.n = self.n + 1

    def resize(self, new_csize):
        # create a new array with new size
        B = self.create_array(new_csize)
        # copy elements of A to B
        for i in range(self.n):
            B[i] = self.A[i]
        # reassign A and update size
        self.A = B
        self.size = new_csize

    # print logic: print list(l)
    def __str__(self):
        result = ' '
        for i in range(self.n):
            result = result + str(self.A[i]) + ','

        return '['+result[:-1]+']'

    # get element by index
    def __getitem__(self, index):
        if 0 <= index < self.n:
            return self.A[index]
        else:
            return IndexError("Index out of range")

    # custom method to get element by index
    def indexing(self, index):
        if 0 <= index < self.n:
            return self.A[index]
        else:
            return IndexError("Index out of range")

    #delete last element
    def pop(self):
        if self.n ==0:
            return 'empty list'
        
        print(self.A[self.n-1])
        self.n = self.n -1

    #clear whole list
    def clear(self):
        self.n = 0
        self.size = 1

    #find element element postition
    def find(self, item):
        for i in range(self.n):
            if self.A[i]==item:
                return i 

        return "value error: not in list"

    #insert element in particular index

    def insert(self, index, element):
        if index < 0 or index > self.n:
            return 'invalid index'

        # if array is full, resize
        if self.n == self.size:
            self.resize(self.size * 2)

        # shift elements to the right
        for i in range(self.n, index, -1):
            self.A[i] = self.A[i-1]

        # insert new element
        self.A[index] = element
        self.n += 1
      
    #delete element in particular index
    
    def delete(self, index):
        if index < 0 or index >= self.n:
            return 'invalid index'

        # shift elements to the left
        for i in range(index, self.n - 1):
            self.A[i] = self.A[i + 1]

        # clear the last slot (optional)
        self.A[self.n - 1] = None  

        # reduce count
        self.n -= 1

    def remove(self, element):
    # search for the element
        for i in range(self.n):
            if self.A[i] == element:
                # shift everything left
                for j in range(i, self.n - 1):
                    self.A[j] = self.A[j + 1]

                # clear last slot (optional)
                self.A[self.n - 1] = None
                self.n -= 1
                return  # exit after first removal

        # if element not found
        return "element not found"

    def remove(self, element):
        pos = self.find(element)   # use find() to get index
        if isinstance(pos, int):   # found
            # shift left
            for i in range(pos, self.n - 1):
                self.A[i] = self.A[i + 1]

            self.A[self.n - 1] = None  # clear last slot
            self.n -= 1
        else:
            return pos   # "value error: not in list"

```
