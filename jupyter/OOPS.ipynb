```python
class Employee:
    def __init__(self, name=None, empid=None, salary=None):
        self.name = name
        self.empid = empid
        self.salary = salary

    def setempid(self, empid):
        self.empid = empid

    def setname(self, name):
        self.name = name

    def setsalary(self, salary):
        self.salary = salary

    def getempid(self):
        return self.empid

    def getname(self):
        return self.name

    def getsalary(self):
        return self.salary


e1 = Employee()
e2 = Employee('rahul', 11, 126789)

e1.setempid(22)
e1.setname("robin")
e1.setsalary(12345678)

print(e1.getempid(), e1.getname(), e1.getsalary())
print(e2.getempid(), e2.getname(), e2.getsalary())
```
