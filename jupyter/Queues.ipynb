```python
class Node:
    def __init__(self, value):
        self.data = value
        self.next = None
```

```python

class queue:
    def __init__(self):
        self.front = None
        self.rear = None

    def is_empty(self):
        """Check if the queue is empty"""
        return self.front is None

    def enqueue(self, value):
        new_node = Node(value)

        if self.rear is None:  # Queue is empty
            self.front = new_node
            self.rear = new_node
        else:
            self.rear.next = new_node
            self.rear = new_node

    def dequeue(self):
        if self.is_empty():
            print("Queue is empty ⚠️")
            return None
        data = self.front.data
        self.front = self.front.next

        # If queue becomes empty after dequeuing, reset rear
        if self.front is None:
            self.rear = None

        return data

    def traverse(self):
        if self.is_empty():
            print("Queue is empty ⚠️")
            return
        curr = self.front
        print("Queue elements:", end=" ")
        while curr is not None:
            print(curr.data, end=" ")
            curr = curr.next
        print()

    def size(self):
        temp = self.front
        count = 0

        while temp != None:
            count += 1
            temp = temp.next

        return count
    
    def front_item(self):
        if self.front == None:
            return 'empty'
        
        else:
            return self.front.data
        
    
    def rear_item(self):
        if self.rear == None:
            return "empty"

        else:
            return self.rear.data
        
```

```python
q = queue()
```

```python
q.enqueue(2)
q.enqueue(3)
q.enqueue(4)
q.enqueue(5)
print(q)
```

```python
q.traverse()           # Queue elements: 10 20 30
```

```python
print("Is empty?", q.is_empty())  # False

print("Dequeued:", q.dequeue())  # Dequeued: 10
q.traverse()
```

```python
def fun(n, q):
    if n == 0:
        return 0
    else:
        q.enqueue(n % 10)       # enqueue last digit
        res = fun(n // 10, q)   # recursive call on remaining digits
        res = res * 10 + q.dequeue()  # dequeue in reverse order
        return res


# --- Test ---
q1 = queue()
print(fun(123, q1))  # Expected output: 321
```
