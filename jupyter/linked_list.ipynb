```python
class _linked_list:

    def __init__(self):
        
        #Empty linked list
        self.head = None # first node: always none
        self.n = 0 #n : use for count no. of linkedlsit

     #length of linked_list   
    def __len__(self):
        return self.n
    
    #insert
        #head insert :  
        #tail insert :append methord
        #midle insert : insert methord

    def insert_head(self, value):
        #create new node
        new_node = Node(value)

        #new node address store in current head
        new_node.next = self.head
        
        # now new node is head
        self.head = new_node

        # counting also increase 
        self.n = self.n + 1
        
    # insert at last of linked_list
    def insert_tail(self, value):
        new_node = Node(value)

        if self.head == None:
            self.head = new_node #empty head
            self.n = self.n +1
            return

        curr = self.head

        while curr.next != None:
            curr = curr.next

        curr.next = new_node
        self.n = self.n +1

    #insert_after: meaning is insert a value after a element/Node
    def inser_after(self, after, value):

        new_node = Node(value)
        curr = self.head
        
        while curr != None:
            if curr.data == after:
                break 
            curr = curr.next

        # case 1 break -> item apko mil gaya ->curr -> none
        if curr != None:
            new_node.next = curr.next
            curr.next = new_node
            self.n = self.n + 1
        else:
            return 'Item not found'

        print(curr.data)


    #traverse
        #print
    #1st way
    def traverse(self):
        current = self.head

        while current != None:
            print(current.data)
            current = current.next
    #2nd way
    def __str__(self):
        current = self.head
        result = " "
        while current != None:
            print(current.data)
            result = result + str(current.data) + "->"
            current = current.next
        return result[:-2]
    
    #delete
        #clear
        #head delete
        #tail: pop
        # value: remove
        #index: 
    def clear(self):
        self.head = None
        self.n = 0

    def delete_head(self):
        if self.head == None:
            return "empty"
        
        self.head = self.head.next

    def delete_tail(self):
        
        if self.head == None:
            return "empty"
        
        curr = self.head

        if curr.next == None:
            # only one item: then we have to delete from head
            return self.delete_head()

        while curr.next.next != None:
            curr = curr.next
        
        curr.next = None
        self.n = self.n - 1
            
    def remove_delete_value(self, value):

        #check empty or not: only one node
        if self.head == None:
            return "empty"
        
        # first node value found
        if self.head.data == value:
            return self.delete_head()

        curr = self.head
        while curr.next != None:
            if curr.next.data == value:
                break
            curr = curr.next

        #case 1 : element mil gaya
        #case 2: element ni mila

        if curr.next== None:
            return "not found"
        else:
            curr.next = curr.next.next


    #search
        #value search
        #index search

    def value_search(self, value):
        curr = self.head
        pos = 0

        while curr != None:
            if curr.data == value:
                return pos
            curr = curr.next
            pos = pos + 1

        return " not found "
    
    # def index_search(self, index): # 1st way
    def __getitem__(self, index): #2nd way
        curr = self.head
        pos = 0

        while curr is not None:
            if pos == index:
                return curr.data
            curr = curr.next
            pos += 1

        return "Index out of range"  # or return None, or raise IndexError

            



    #reverce
    #max


```

```python
def sum_odd_pos(self):
    curr = self.head
    counter = 0
    sum_ = 0

    while curr is not None:
        if counter % 2 != 0:   # odd positions (0-based)
            sum_ += curr.data
        counter += 1           # ✅ FIXED increment
        curr = curr.next 

    print(sum_)
    return sum_
```

```python
def reverse(self):
    prev = None
    curr = self.head

    while curr is not None:
        nxt = curr.next       # store next node
        curr.next = prev      # reverse link
        prev = curr           # move prev ahead
        curr = nxt            # move curr ahead

    self.head = prev
    
```

```python
_linked_list.reverse = reverse

l.traverse()
print("------- Reversing -------")
l.reverse()
l.traverse()
```
