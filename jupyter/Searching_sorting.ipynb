```python
#linear
def linear_search(arr, item):
    for i in range(len(arr)):
        if arr[i] == item:
            return i 

    return -1
```

```python
arr = [1,2,3,4,6,7,8]
linear_search(arr, 4)
```

```python
#Binary search

def binary_search(arr, low, high, item):

    print("low =", low, "high =", high, end = ' ')
    if low <= high:
        mid = (low + high)//2

        print("mid value is : ", mid )

        if arr[mid] == item:
            return mid
        elif arr[mid] > item:
            return binary_search(arr, low, mid-1, item)
        else:
            return binary_search(arr, mid+1, high, item)

    else: 
        return -1


```

```python
arr = [12,23,53,65,76,82, 87,93]
print(binary_search(arr, 0, len(arr)-1, 200))
```

```python
# sorting: 
print("how to detected given error is sorted or not sorted")

def is_sorted(arr):
    sorted = True

    for i in range(len(arr) -1):
        if arr[-1]>arr[i+1]:
            sorted = False

    return sorted
```

```python
arr = [12,4,6,7,3,3,6]
is_sorted(arr)
```

```python
def bubble_sort(arr):
    for i in range(len(arr)):
        for j in range(i + 1, len(arr)):
            # keep your style: compare i and j first
            if i < j:
                # bubble-sort style comparison using j-1 and j
                if arr[j - 1] > arr[j]:
                    arr[j - 1], arr[j] = arr[j], arr[j - 1]
            elif i > j:
                pass
            elif i == j:
                pass
    return arr

print(bubble_sort([5, 2, 9, 1, 5, 6]))

#clean code

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        # Last i elements are already in place
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

print(bubble_sort([5, 2, 0, 1, 5, 6]))

```

```python
def bubble_sort(arr):
    comparisons = 0
    swaps = 0
    
    n = len(arr)

    for i in range(n):
        for j in range(0, n - i - 1):

            comparisons += 1
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swaps += 1

    return arr, comparisons, swaps

sorted_arr, comps, sws = bubble_sort([5, 2, 9, 1])
print("Sorted:", sorted_arr)
print("Comparisons:", comps)
print("Swaps:", sws)
```

```python
#Bubble sort : By nature is not adaptive BUUT we can make by using flag
def bubble(arr):
    n = len(arr)
    
    for i in range(n):
        flag = 0
        
        for j in range(0, n - 1 - i):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                flag = 1
        
        # If no swaps happened, array is already sorted → stop early
        if flag == 0:
            break

    print(arr)
```

```python
arr = [1,2,34,6,6]
bubble(arr)
```

```python
#selection sort
print("SELECTION SORT: \n is Adaptable : False \n is Statble : True")

def selection_sort(arr):
    n = len(arr)

    for i in range(n):
        min_index = i

        for j in range(i + 1, n):
            if arr[j] < arr[min_index]:
                min_index = j

        arr[i], arr[min_index] = arr[min_index], arr[i]

    print(arr)
```

```python
arr = [1,2,34,6,6]
selection_sort(arr)
```

```python
def merge_sort(arr):
    mid = len(arr) // 2
    left = arr[:mid]
    right = arr[mid:]
    return left, right

def merge(left, right):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result

```

```python
print(merge_sort([5, 2, 9, 1, 3, 7]))
# Output: [1, 2, 3, 5, 7, 9]
```
