```python
print("linked_list STACK")
class node: 
    def __init__(self, value):
        self.data = value
        self.next = None
        
```

```python
class stack:

    def __init__(self):
        self.top = None

    def isempty(self):
        return self.top is None
    
    def push(self, value):
        new_node = node(value)
        new_node.next = self.top
        self.top = new_node

    def pop(self):
        if self.top is None:
            print("Stack underflow — nothing to pop")
            return None
        popped = self.top.data
        self.top = self.top.next
        return popped

    def peek(self):
        if(self.isempty()):
            return "stack empty"
        else: 
            return self.top.data

    def traverse(self):
        curr = self.top
        if curr is None:
            print("Stack is empty.")
            return
        while curr is not None:
            print(curr.data)
            curr = curr.next

    def size(self):
        return self.count
```

```python
def find_the_celeb(l):
    s = stack()

    # Step 1: push everyone onto stack
    for i in range(len(l)):
        s.push(i)

    # Step 2: narrow down to one potential celeb
    while s.size() >= 2:
        i = s.pop()
        j = s.pop()

        if l[i][j] == 1:
            # i knows j → i cannot be celeb
            s.push(j)
        else:
            # i doesn't know j → j cannot be celeb
            s.push(i)

    celeb = s.pop()

    # Step 3: verify celeb
    for i in range(len(l)):
        if i != celeb:
            if l[i][celeb] == 0 or l[celeb][i] == 1:
                print("No celebrity found.")
                return

    print("The celebrity is:", celeb)
```

```python
# ques: balance parnathese
def balance_parenthesis(expr):
    s = stack()
    pairs = {')': '(', ']': '[', '}': '{'}

    for ch in expr:
        if ch in ['(', '[', '{']:
            s.push(ch)
        elif ch in [')', ']', '}']:
            if s.isempty() or s.pop() != pairs[ch]:
                print("Not balanced ❌")
                return

    if s.isempty():
        print("Balanced ✅")
    else:
        print("Not balanced ❌")
```

```python

# --- Test cases ---
balance_parenthesis("{[()]}")
balance_parenthesis("{[(])}")
balance_parenthesis("((()))")
balance_parenthesis("{[}")
```

```python
print("ARRAY STACK")
class array_stack: 
    def __init__(self, size):
        self.size = size
        self.stack = [None] * self.size
        self.top = -1 

    def _push(self, value):
        if self.top == self.size - 1:
            return "Overflow"
        else:
            self.top += 1
            self.stack[self.top] = value  

    def pop(self):
        if self.top == -1:
            return "empty"
        else:
            data = self.stack[self.top]
            self.top -= 1
            print(data)

    def traverse(self):
        if self.top == -1:
            print("Stack is empty ⚠️")
        else:
            print("Stack elements:", end=" ")
            for i in range(self.top + 1):
                print(self.stack[i], end=" ")
```
