```python
class Dictionary:
    def __init__(self, size):
        self.size = size
        self.slots = [None] * self.size  # keys
        self.data = [None] * self.size   # values

    def hash_function(self, key):
        """Simple hash function."""
        return abs(hash(key)) % self.size

    def rehash(self, old_hash):
        """Linear probing step."""
        return (old_hash + 1) % self.size

    def put(self, key, value):
        """Insert or update a key-value pair."""
        hash_value = self.hash_function(key)

        if self.slots[hash_value] is None:
            # Empty slot found
            self.slots[hash_value] = key
            self.data[hash_value] = value
        elif self.slots[hash_value] == key:
            # Key already exists → update
            self.data[hash_value] = value
        else:
            # Collision → rehash
            new_hash = self.rehash(hash_value)
            while self.slots[new_hash] is not None and self.slots[new_hash] != key:
                new_hash = self.rehash(new_hash)

            # Insert or update
            self.slots[new_hash] = key
            self.data[new_hash] = value

    def get(self, key):
        """Retrieve value for a key."""
        start_pos = self.hash_function(key)
        pos = start_pos

        while self.slots[pos] is not None:
            if self.slots[pos] == key:
                return self.data[pos]
            pos = self.rehash(pos)
            if pos == start_pos:
                break  # full cycle

        return None  # not found

    def __setitem__(self, key, value):
        self.put(key, value)

    def __getitem__(self, key):
        return self.get(key)

    def __str__(self):
        """Print dictionary contents."""
        for i in range(self.size):
            if self.slots[i] is not None:
                print(f"{self.slots[i]} : {self.data[i]}")
        return ""
```

```python
# Chaining Hash Table with Traverse
class HashTable:
    def __init__(self, size):
        self.size = size
        self.table = [[] for _ in range(size)]  # each slot stores a list of (key, value) pairs

    def hash_function(self, key):
        """Simple hash function."""
        return abs(hash(key)) % self.size

    def insert(self, key, value):
        """Insert or update a key-value pair."""
        hash_value = self.hash_function(key)
        bucket = self.table[hash_value]

        # Check if key already exists → update value
        for pair in bucket:
            if pair[0] == key:
                pair[1] = value
                print(f"Updated key '{key}' with new value '{value}' at index {hash_value}")
                return

        # Otherwise, append new pair
        bucket.append([key, value])
        print(f"Inserted key '{key}' with value '{value}' at index {hash_value}")

    def search(self, key):
        """Search for a key and return its value."""
        hash_value = self.hash_function(key)
        bucket = self.table[hash_value]

        for pair in bucket:
            if pair[0] == key:
                print(f"Found key '{key}' at index {hash_value} with value '{pair[1]}'")
                return pair[1]

        print(f"Key '{key}' not found.")
        return None

    def delete(self, key):
        """Delete a key-value pair from the hash table."""
        hash_value = self.hash_function(key)
        bucket = self.table[hash_value]

        for i, pair in enumerate(bucket):
            if pair[0] == key:
                del bucket[i]
                print(f"Deleted key '{key}' from index {hash_value}")
                return True

        print(f"Key '{key}' not found for deletion.")
        return False

    def display(self):
        """Display the entire hash table with chains."""
        print("\nHash Table Contents:")
        for i in range(self.size):
            print(f"Index {i}: {self.table[i]}")

    def traverse(self):
        """Traverse through all key-value pairs in the hash table."""
        print("\nTraversing all elements in the hash table:")
        for i in range(self.size):
            if len(self.table[i]) > 0:
                for pair in self.table[i]:
                    print(f"Index {i} → Key: {pair[0]}, Value: {pair[1]}")
            else:
                print(f"Index {i} → Empty")
```

```python
# Node for Linked List
class Node:
    def __init__(self, key, value):
        self.key = key        # key (like a word)
        self.value = value    # value (like meaning)
        self.next = None      # pointer to next node


# Linked List class for chaining
class LinkedList:
    def __init__(self):
        self.head = None

    def insert(self, key, value):
        """Insert or update a key-value pair in the linked list."""
        current = self.head
        while current:
            if current.key == key:
                current.value = value  # update existing key
                print(f"Updated key '{key}' with value '{value}' in linked list.")
                return
            current = current.next

        # key not found → insert new node at beginning
        new_node = Node(key, value)
        new_node.next = self.head
        self.head = new_node
        print(f"Inserted key '{key}' with value '{value}' in linked list.")

    def search(self, key):
        """Search for a key and return its value."""
        current = self.head
        while current:
            if current.key == key:
                print(f"Found key '{key}' with value '{current.value}' in linked list.")
                return current.value
            current = current.next
        print(f"Key '{key}' not found in linked list.")
        return None

    def delete(self, key):
        """Delete a node by its key."""
        current = self.head
        previous = None

        while current:
            if current.key == key:
                if previous:
                    previous.next = current.next
                else:
                    self.head = current.next
                print(f"Deleted key '{key}' from linked list.")
                return True
            previous = current
            current = current.next

        print(f"Key '{key}' not found for deletion.")
        return False

    def display(self):
        """Display the linked list."""
        current = self.head
        while current:
            print(f"({current.key}: {current.value})", end=" -> ")
            current = current.next
        print("None")


# Hash Table class using Linked List chaining
class HashTable:
    def __init__(self, size):
        self.size = size
        self.table = [LinkedList() for _ in range(size)]

    def hash_function(self, key):
        """Simple hash function."""
        return abs(hash(key)) % self.size

    def insert(self, key, value):
        """Insert or update a key-value pair in the hash table."""
        hash_index = self.hash_function(key)
        print(f"\nInserting '{key}' at index {hash_index}")
        self.table[hash_index].insert(key, value)

    def search(self, key):
        """Search for a key in the hash table."""
        hash_index = self.hash_function(key)
        print(f"\nSearching for '{key}' at index {hash_index}")
        return self.table[hash_index].search(key)

    def delete(self, key):
        """Delete a key from the hash table."""
        hash_index = self.hash_function(key)
        print(f"\nDeleting '{key}' at index {hash_index}")
        return self.table[hash_index].delete(key)

    def display(self):
        """Display the whole hash table."""
        print("\nHash Table Contents:")
        for i in range(self.size):
            print(f"Index {i}: ", end="")
            self.table[i].display()
```

```python
# Create a hash table of size 5
h = HashTable(5)

# Insert some values
h.insert("apple", 100)
h.insert("banana", 200)
h.insert("grape", 300)
h.insert("orange", 400)
h.insert("melon", 500)

# Show table
h.display()

# Search for some keys
h.search("banana")
h.search("mango")

# Delete a key
h.delete("apple")
h.display()
```

```python
# Node for Linked List
class Node:
    def __init__(self, key, value):
        self.key = key        # key (like a word)
        self.value = value    # value (like meaning)
        self.next = None      # pointer to next node


# Linked List class for chaining
class LinkedList:
    def __init__(self):
        self.head = None

    def insert(self, key, value):
        """Insert or update a key-value pair in the linked list."""
        current = self.head
        while current:
            if current.key == key:
                current.value = value  # update existing key
                print(f"Updated key '{key}' with value '{value}' in linked list.")
                return
            current = current.next

        # key not found → insert new node at beginning
        new_node = Node(key, value)
        new_node.next = self.head
        self.head = new_node
        print(f"Inserted key '{key}' with value '{value}' in linked list.")

    def search(self, key):
        """Search for a key and return its value."""
        current = self.head
        while current:
            if current.key == key:
                print(f"Found key '{key}' with value '{current.value}' in linked list.")
                return current.value
            current = current.next
        print(f"Key '{key}' not found in linked list.")
        return None

    def delete(self, key):
        """Delete a node by its key."""
        current = self.head
        previous = None

        while current:
            if current.key == key:
                if previous:
                    previous.next = current.next
                else:
                    self.head = current.next
                print(f"Deleted key '{key}' from linked list.")
                return True
            previous = current
            current = current.next

        print(f"Key '{key}' not found for deletion.")
        return False

    def get_all_pairs(self):
        """Return all key-value pairs in this linked list."""
        pairs = []
        current = self.head
        while current:
            pairs.append((current.key, current.value))
            current = current.next
        return pairs

    def display(self):
        """Display the linked list."""
        current = self.head
        while current:
            print(f"({current.key}: {current.value})", end=" -> ")
            current = current.next
        print("None")


# Hash Table class using Linked List chaining with Rehashing
class HashTable:
    def __init__(self, size):
        self.size = size
        self.count = 0               # how many key-value pairs total
        self.table = [LinkedList() for _ in range(size)]
        self.load_factor_limit = 0.7 # rehash when load factor exceeds 0.7

    def hash_function(self, key):
        """Simple hash function."""
        return abs(hash(key)) % self.size

    def insert(self, key, value):
        """Insert or update a key-value pair in the hash table."""
        hash_index = self.hash_function(key)
        print(f"\nInserting '{key}' at index {hash_index}")
        before = self.count

        # Check if key already exists in that chain
        current = self.table[hash_index].head
        while current:
            if current.key == key:
                current.value = value
                print(f"Updated key '{key}' with value '{value}' (no new element).")
                return
            current = current.next

        # Insert new element
        self.table[hash_index].insert(key, value)
        self.count += 1

        # Check load factor and rehash if necessary
        if self.load_factor() > self.load_factor_limit:
            print("\n⚠️ Load factor exceeded limit. Rehashing table...\n")
            self.rehash()

    def search(self, key):
        """Search for a key in the hash table."""
        hash_index = self.hash_function(key)
        print(f"\nSearching for '{key}' at index {hash_index}")
        return self.table[hash_index].search(key)

    def delete(self, key):
        """Delete a key from the hash table."""
        hash_index = self.hash_function(key)
        print(f"\nDeleting '{key}' at index {hash_index}")
        deleted = self.table[hash_index].delete(key)
        if deleted:
            self.count -= 1
        return deleted

    def load_factor(self):
        """Compute the current load factor."""
        return self.count / self.size

    def rehash(self):
        """Resize the table and reinsert all elements."""
        old_table = self.table
        old_size = self.size
        self.size = self.size * 2  # double the size
        self.table = [LinkedList() for _ in range(self.size)]
        self.count = 0

        # Reinsert all elements
        for i in range(old_size):
            pairs = old_table[i].get_all_pairs()
            for key, value in pairs:
                self.insert(key, value)

        print(f"✅ Rehashing complete. New table size: {self.size}")

    def display(self):
        """Display the whole hash table."""
        print("\nHash Table Contents:")
        for i in range(self.size):
            print(f"Index {i}: ", end="")
            self.table[i].display()

    def traverse(self):
        """Traverse all elements."""
        print("\nTraversing all elements in the hash table:")
        for i in range(self.size):
            current = self.table[i].head
            while current:
                print(f"Index {i} → Key: {current.key}, Value: {current.value}")
                current = current.next
```
