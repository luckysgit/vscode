````python
def reverse_string(st):
    rev = ""
    for i in range(len(st)-1, -1, -1):
        rev += st[i]
    return rev
````

# 🔤 **String Question 2 (Easy)**

### **Problem: Check if a String is a Palindrome**

A string is a **palindrome** if it reads the same forward and backward.

### **Example**

Input:

```
"madam"
```

Output:

```
True
```

Input:

```
"hello"
```

Output:

```
False
```

---

# ⭐ Your Task:

1. Explain how to check if a string is a palindrome
2. Write the Python code

Take your time — your turn!


````python
def pal(st):
    if st[:: -1] == st:
        return True
    else:
        return False
````

````python
st = "mdam"
pal(st)
````

# 🔤 **String Question 3 (Easy)**

### **Problem: Count Vowels in a String**

Given a string `s`, return the **number of vowels** in it.
Vowels are: `a, e, i, o, u` (both lowercase + uppercase)

### **Example**

Input:

```
"Hello World"
```

Output:

```
3
```

(‘e’, ‘o’, ‘o’)

---

# ⭐ Your Task:

1. Explain how you will count vowels
2. Write the Python code

---

Your turn!


````python
def count_vowels(st):
    vowels = ['a', 'e', 'i', 'o', 'u']
    count = 0
    
    for ch in st.lower():      # lowercase to handle 'A', 'E'...
        if ch in vowels:
            count += 1
    
    return count
````

````python
print(count_vowels("Hello World"))
````
