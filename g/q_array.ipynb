````python
def longest_consecutive(nums):
    num_set = set(nums)
    longest = 0

    for n in nums:
        # only start at sequence beginning
        if n - 1 not in num_set:
            current = n
            length = 1

            while current + 1 in num_set:
                current += 1
                length += 1

            longest = max(longest, length)

    return longest
````

````python
a = [100, 4, 200, 1, 3, 2]
longest_consecutive(a)
````

# 🔥 **Array Question 4 (Medium-Hard)**

### **Problem: Kadane’s Algorithm — Maximum Subarray Sum**

Given an integer array `nums`, find the **maximum possible sum** of any **contiguous subarray**.

### **Example**

Input:

```
nums = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
```

Output:

```
6
```

### **Explanation**

The subarray with the maximum sum is:

```
[4, -1, 2, 1]  → sum = 6
```

---

# ⭐ Your Task:

1. Explain your approach
2. Write Python code for it

   * You can use either
     ✔ Brute force (O(n²))
     ✔ Kadane’s algorithm (O(n)) — preferred by Google

---

Whenever ready, give your explanation + code!


````python
def max_subarray(arr):
    max_sum = arr[0]
    current_sum = 0

    for num in arr:
        current_sum += num
        
        if current_sum > max_sum:
            max_sum = current_sum
        
        if current_sum < 0:
            current_sum = 0

    return max_sum


nums = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
print(max_subarray(nums))
````

# ⭐ Your Task:

1. Explain how you will find the missing number
2. Write the Python code

You may use:

* ✔ Sum formula method → O(n), no extra space
* ✔ XOR method → O(n), no extra space (Google loves this)
* ✔ Hashing (but uses extra space)

---

Take your time.

````python
def missing_n(arr):
    n = len(arr)
    expected_sum = n * (n + 1) // 2
    actual_sum = sum(arr)
    return expected_sum - actual_sum
````

````python
print(missing_n([3, 0, 1]))  
````

````python
def missing(arr):
    n = len(arr)
    xor1 = 0
    xor2 = 0
    
    for i in range(n):
        xor1 ^= arr[i]
        xor2 ^= i
    
    xor2 ^= n
    
    return xor1 ^ xor2
````

````python
print(missing([3, 0, 1]))  
````

# 🔥 **Array Question 6 (Hard — Google Top Interview Question)**

### **Problem: Merge Intervals**

You are given an array of intervals where each interval is a pair:

```
[ start, end ]
```

Merge all overlapping intervals and return the result.

---

### **Example**

Input:

```
intervals = [[1,3],[2,6],[8,10],[15,18]]
```

Output:

```
[[1, 6], [8, 10], [15, 18]]
```

### **Explanation**

* `[1,3]` and `[2,6]` overlap → merge into `[1,6]`
* Others do not overlap

---

# ⭐ Your Task:

1. Explain your approach
2. Write Python code to solve it

Hint:
✔ First **sort** intervals by starting time
✔ Then merge by checking overlapping ranges

---

Whenever you're ready, give your **approach + code**!


````python
def merge_intervals(intervals):
    intervals.sort()  # sort by first element
    
    merged = []
    
    for interval in intervals:
        if not merged or merged[-1][1] < interval[0]:
            # no overlap → add interval
            merged.append(interval)
        else:
            # overlap → merge
            merged[-1][1] = max(merged[-1][1], interval[1])
    
    return merged
````

````python
print(merge_intervals([[1,3],[2,6],[8,10],[15,18]]))
````
